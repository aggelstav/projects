```python
import tsfresh
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset2017.csv", index_col=0, header=[1])
df = df .dropna()
df.head()
```

```python
#suppose with want to make a prediction for "4BLTE"
# its index is 8

def split_sequences(sequences, n_steps, idx_array):
    X, target = list(), list()
    for i in range(len(sequences)):
        #find the end of the subsequence
        end_ix = i + n_steps
        if end_ix > len(sequences) - 1:
            break
        #gather input and output parts of the subsequence
        seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix, :]
        target.append(np.where(seq_x[-1, idx_array]<seq_y[idx_array], 1, 0))
        X.append(seq_x)
    return np.array(X), np.array(target)
        
```

```python
X, y = split_sequences(df.values, 2, 5)
y = pd.Series(y)
```

```python
X = tsfresh.extract_relevant_features(df, y,column_id='1CLTE')
```

```python
from tsfresh.examples.robot_execution_failures import download_robot_execution_failures, \
    load_robot_execution_failures
download_robot_execution_failures()
timeseries, y = load_robot_execution_failures()
```

```python
print(timeseries.head(5))
import matplotlib.pyplot as plt
timeseries[timeseries['id'] == 3].plot(subplots=True, sharex=True, figsize=(10,10))
plt.show()
```

```python
timeseries[timeseries['id'] == 21].plot(subplots=True, sharex=True, figsize=(10,10))
plt.show()
```

```python
from tsfresh import extract_features
extracted_features = extract_features(timeseries, column_id="id", column_sort="time")
```

```python
from tsfresh import select_features
from tsfresh.utilities.dataframe_functions import impute

impute(extracted_features)
features_filtered = select_features(extracted_features, y)
features_filtered
```

```python
from tsfresh import extract_relevant_features

features_filtered_direct = extract_relevant_features(timeseries, y,
                                                     column_id='id', column_sort='time')
```

```python
from xgboost import XGBClassifier, plot_importance, plot_tree
model = XGBClassifier(n_jobs=-1)
```

```python
model.fit(features_filtered_direct,y)
ax = plot_importance(model)
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
kf = KFold(n_splits=10)

model = XGBClassifier(nthread=-1)
results = cross_val_score(model, features_filtered_direct, y, cv=kf, scoring='accuracy', n_jobs=-1)
```

```python
def evaluate_score(score):
    mean = np.mean(score)
    std = np.std(score)
    print(f"Average score:{mean}+/-{std}%")
```

```python
evaluate_score(results)
```
